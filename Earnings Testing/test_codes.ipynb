```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
from pandas.io.json import json_normalize
import urllib.request as urlreq
import json
import datetime

quick_summary_url = 'https://query1.finance.yahoo.com/v7/finance/quote?symbols={}'
```

```python
earnings_url = 'https://finance.yahoo.com/calendar/earnings?day={0}&offset={1}&size=100'

dt = '2019-11-14'

earnings_soup = BeautifulSoup(requests.get(earnings_url.format(dt, 0)).text, 'lxml')

earnings_soup = earnings_soup.find("div", {"id": "fin-cal-table"})

total_names = int(earnings_soup.find('span', {'data-reactid':'8'}).text.replace('results','').strip().split(' ')[-1])
```

```python
def earnings_table(dt, offset_amount):
    
    earnings_url = 'https://finance.yahoo.com/calendar/earnings?day={0}&offset={1}&size=100'
    
    earnings_soup = BeautifulSoup(requests.get(earnings_url.format(dt, offset_amount)).text, 'lxml')
    
    earnings_soup = earnings_soup.find("div", {"id": "fin-cal-table"})
    
    curr_table = earnings_soup.find('table', {'data-reactid': '17'})

    headers = [th.text for th in curr_table.find('tr', {'data-reactid': '19'}).find_all('th')]

    curr_dict = {}

    for header in headers:
        curr_dict[header] = []

    for row in curr_table.find('tbody').find_all('tr'):
        curr_row_text = [td.text for td in row.find_all('td')]

        for i, val in enumerate(curr_row_text):
            curr_dict[headers[i]].append(val)
```

```python
init_df = pd.DataFrame(curr_dict)
```

```python
earnings_url = 'https://finance.yahoo.com/calendar/earnings?day={0}&offset={1}&size=100'
    
earnings_soup = BeautifulSoup(requests.get(earnings_url.format(dt, 100)).text, 'lxml')

earnings_soup = earnings_soup.find("div", {"id": "fin-cal-table"})

curr_table = earnings_soup.find('table', {'data-reactid': '17'})

headers = [th.text for th in curr_table.find('tr', {'data-reactid': '19'}).find_all('th')]

curr_dict = {}

for header in headers:
    curr_dict[header] = []

for row in curr_table.find('tbody').find_all('tr'):
    curr_row_text = [td.text for td in row.find_all('td')]

    for i, val in enumerate(curr_row_text):
        curr_dict[headers[i]].append(val)
```

```python
earnings_df = pd.concat([init_df, pd.DataFrame(curr_dict)], axis = 0).drop_duplicates()
```

```python
tickers = earnings_df['Symbol'].tolist()

row_lst = []

for ticker in tickers:
    try:
        with urlreq.urlopen(quick_summary_url.format(ticker)) as url:
            data = json.loads(url.read().decode())

        curr_df = pd.DataFrame(data['quoteResponse']['result'][0], index = [ticker])
        row_lst.append(curr_df)
    except:
        continue
    time.sleep(1)
```

```python
earnings_info = pd.concat(row_lst, axis = 0)
earnings_info = earnings_info[earnings_info.index.isin(earnings_df[earnings_df['Earnings Call Time'] == 'After Market Close'].Symbol.tolist())]
```

```python
earnings_info[['ask', 'bid', 'bidSize', 'bookValue', 'currency', 'dividendDate',
          'forwardPE', 'marketCap','priceToBook','regularMarketChange',
          'regularMarketChangePercent', 'regularMarketDayHigh',
       'regularMarketDayLow', 'regularMarketDayRange', 'regularMarketOpen',
       'regularMarketPreviousClose', 'regularMarketPrice', 
         'regularMarketVolume', 'sharesOutstanding', 'shortName',
         'trailingPE', 'twoHundredDayAverage']].to_csv('earnings.csv')
```

```python
options_url = 'https://query1.finance.yahoo.com/v7/finance/options/{}'
```

```python
ticker = 'NVDA'

options = json.loads(requests.get(options_url.format(ticker)).text)

options = options['optionChain']['result'][0]['options'][0]
expiry = datetime.datetime.utcfromtimestamp(int(options['expirationDate'])).strftime('%Y-%m-%d')

calls_json = options['calls']
puts_json = options['puts']

calls = {'strike': [],
         'bid': [],
         'ask': [],
         'iv': [],
         'expiration': []}

for call in calls_json:
    calls['strike'].append(float(call['strike']))
    calls['bid'].append(float(call['bid']))
    calls['ask'].append(float(call['ask']))
    calls['iv'].append(float(call['impliedVolatility']))
    calls['expiration'].append(datetime.datetime.utcfromtimestamp(int(call['expiration'])))
    
calls = pd.DataFrame(calls)

puts = {'strike': [],
         'bid': [],
         'ask': [],
         'iv': [],
         'expiration': []}

for put in puts_json:
    puts['strike'].append(float(put['strike']))
    puts['bid'].append(float(put['bid']))
    puts['ask'].append(float(put['ask']))
    puts['iv'].append(float(put['impliedVolatility']))
    puts['expiration'].append(datetime.datetime.utcfromtimestamp(int(put['expiration'])))
    
puts = pd.DataFrame(puts)

ivs = puts[['strike','iv']].set_index('strike')
ivs.columns = ['put_iv']
ivs = ivs.join(calls[['strike','iv']].set_index('strike'))
ivs.columns = ['put_iv','call_iv']
ivs.plot(figsize = (20,10))
```
