```python
import wikipedia as wiki
import pandas as pd
import numpy as np
import datetime as dt
from bs4 import BeautifulSoup as bs
pd.options.display.float_format = '{:,.4f}'.format
```

```python
wiki_content = wiki.page('List_of_accidents_and_incidents_involving_the_Boeing_737')
wiki_html = bs(wiki_content.html(),'html')
```

```python
ba_crashes = pd.DataFrame(columns = ['AccidentDate', 'Description'])

i = 0
for list_item in wiki_html.find_all('li'):
    
    if len(list_item.text.split(' – ')) > 1:
        curr_point = list_item.text.split(' – ')
        accident_date = curr_point[0]
        description = curr_point[1]
        try:
            ba_crashes.loc[i , 'AccidentDate'] = dt.datetime.strptime(accident_date, '%B %d, %Y')
            ba_crashes.loc[i , 'Description'] = 'Crash'
        except:
            print(curr_point)
    elif len(list_item.text.split(': ')) > 1:
        curr_point = list_item.text.split(': ')
        accident_date = curr_point[0]
        description = curr_point[1]
        try:
            ba_crashes.loc[i , 'AccidentDate'] = dt.datetime.strptime(accident_date, '%B %d, %Y')
            ba_crashes.loc[i , 'Description'] = 'Crash'
        except:
            try:
                ba_crashes.loc[i , 'AccidentDate'] = dt.datetime.strptime(accident_date.split(' (')[0], '%B %d, %Y')
                ba_crashes.loc[i , 'Description'] = 'Crash'
            except:
                print(curr_point)
        
    else:
        None
    
    i += 1
print(i)
```

```python
ba_crashes.loc[i + 1, 'AccidentDate'] = dt.datetime(2018, 8, 16)
ba_crashes.loc[i + 1, 'Description'] = 'Crash'

ba_crashes.loc[i + 2, 'AccidentDate'] = dt.datetime(2018, 9, 28)
ba_crashes.loc[i + 2, 'Description'] = 'Crash'

ba_crashes.loc[i + 3, 'AccidentDate'] = dt.datetime(2019, 3, 10)
ba_crashes.loc[i + 3, 'Description'] = 'Crash'
```
