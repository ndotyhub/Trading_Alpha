```python
starting_date = dt.datetime(1990,1,1)

tlt = yahoo_query('TLT', starting_date)
tlt.hist_prices_query()
tlt = tlt.hist_prices

spx = yahoo_query('^GSPC', starting_date)
spx.hist_prices_query()
spx = spx.hist_prices

tlt_lookthru = pd.read_csv('https://www.blackrock.com/investing/products/239454/ishares-20-year-treasury-bond-etf/1464253357814.ajax?fileType=csv', skiprows = 9).dropna().reset_index()
tlt_lookthru.columns = tlt_lookthru.iloc[0]
tlt_lookthru = tlt_lookthru.reindex(tlt_lookthru.index.drop(0))
tlt_lookthru.columns = [x.replace(' (%)', '') for x in tlt_lookthru.columns]
numeric_columns = ['Weight','Price','Market Value','Notional Value','Coupon','YTM','Yield to Worst','Duration','Par Value']

for col in numeric_columns:
    tlt_lookthru[col] = pd.to_numeric(tlt_lookthru[col].str.replace(',',''))
    
tlt_lookthru = tlt_lookthru[tlt_lookthru['Asset Class'] == 'Fixed Income'].reset_index(drop = True)

tlt_lookthru['Mod Duration'] = np.nan

for idx, row in tlt_lookthru.iterrows():
    treasury_link = 'https://markets.businessinsider.com/bonds/{}'.format(row.ISIN)
    t_soup = bs(requests.get(treasury_link).text, 'lxml')
    tables = t_soup.find_all("table", class_="table table-small no-margin-bottom table-hover")
    if len(tables) == 1:
        mod_duration = list(filter(lambda x: 'Duration' in x, [str(x) for x in tables[0].find_all('tr')]))[0].split('bold">')[-1]
        mod_duration = float(mod_duration.split('<')[0])
        tlt_lookthru.loc[idx,'Mod Duration'] = mod_duration

tlt_sensitivity = (tlt_lookthru['Mod Duration']*tlt_lookthru['Weight']/100).fillna(0).sum()


tlt_toplevel = pd.read_csv('https://www.blackrock.com/investing/products/239454/ishares-20-year-treasury-bond-etf/1464253357814.ajax?fileType=csv', nrows = 8)
tlt_toplevel.columns = ['Values']
tlt_nav = float(tlt_toplevel.loc['Total Net Assets','Values'].replace(',',''))
tlt_shares = float(tlt_toplevel.loc['Shares Outstanding','Values'].replace(',',''))

fi_equity = tlt.join(spx)[['TLT_adjclose','^GSPC_adjclose']]
fi_equity.columns = ['TLT','SPX']
```
